Define a quantum circuit - `qc = QuantumCircuit(1)`

Add an X gate - `qc.x(0)`

Add an H gate - `qc.h(0)`

Draw the quantum circuit - `qc.draw()`

Find the histogram:

`svsim = Aer.get_backend('statevector_simulator') 
job = execute(qc,svsim,shots=100)
result = job.result()
counts = result.get_counts(qc)
plot_histogram(counts)`
